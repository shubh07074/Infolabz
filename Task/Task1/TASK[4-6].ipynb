```python
import requests
import numpy as np
import pandas as pd
```

```python
url =  "https://api.mfapi.in/mf"
```

```python
response = requests.get(url)
fund_data = response.json()
```

```python
fund_data
```

```python
#4. Total count of mutual funds
print("The total couunt of mutual funds are:",len(fund_data))
```

```python
#5. Allow user to enter mutual fund code . Print fund name. If not available , print mutual fund not found. 
code = int(input("Enter the mutual fund code:"))
code_found = False
for i in fund_data:
    if i["schemeCode"] == code:
        print("The Fund Name is", i["schemeName"], "of the provided code", i["schemeCode"])
        code_found = True
        break

else:
    print("Invalid Code")
```

```python
#6. Allow user to enter name of mutual fund, if available then print found and return its fund code. If not, mutual fund not found. 
name = input("Enter the name of mutual fund: ")
name_found = False
for i in fund_data:
  if i["schemeName"]==name:
    print("Found")
    print("The code of",i["schemeName"],"is",i["schemeCode"])
    name_found = True
    break
else:
  print("Not Found")
```
