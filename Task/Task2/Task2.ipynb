```python
import pandas as pd
import numpy as np
```

```python
#1.  Merge both file's data and store it in all data.
xl1 = 'RESULT1.xlsx'
xl2 = 'RESULT2.xlsx'
```

```python
df1 = pd.read_excel(xl1)
df2 = pd.read_excel(xl2)
```

```python
merged_xl = pd.concat([df1,df2],ignore_index=True)
merged_xl.to_excel("Overall_Resut.xlsx",index=False)
```

```python
#2. Print student entries based on high to low totals.
sorted_xl = merged_xl.sort_values(by='TOTAL', ascending=False)
sorted_xl
```

```python
#7.  Allow user to insert name of student, print total of that student. If student not available print student not found.
name = input("Enter the name of student: ")
for i in merged_xl["NAME"]:
  if i == name:
    print("The searched student",name,"has", merged_xl.loc[merged_xl['NAME'] == name, 'TOTAL'].values[0],"total marks") 
    #merged_xl.loc[merged_xl['NAME'] == name, 'TOTAL'] "it shows following : Before (,comma) = row,after (,comma)= column"
    #merged_xl['NAME'] == name "it suggests the row  where user input name = name in data"
    break
else:
  print("The enterd name",name,"not found")
```

```python
#8. Print average of total of all students.
Average = merged_xl['TOTAL'].mean()
print("Average of total of all students is : ", Average)
```
