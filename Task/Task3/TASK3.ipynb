```python
#2. PRINT GRAPH FROM COVID API AS SHOWN BELOW ( FROM COVID API , DATES VS CASES GRAPH)

import requests
```

```python
url = "https://data.covid19india.org/data.json"
```

```python
response = requests.get(url)
data = response.json()
```

```python
data.keys()
```

```python
cases_time_series = data['cases_time_series']
dates = [entry['date'] for entry in cases_time_series]
daily_cases = [int(entry['dailyconfirmed']) for entry in cases_time_series]
```

```python
plt.figure(figsize=(12, 6))
plt.bar(dates, daily_cases,width=0.8)
plt.title('Daily COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dates, daily_cases)
plt.title('Daily COVID-19 Cases in India')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#3. PRINT GRAPH FROM COVID API AS SHOWN BELOW ( HORIZONTAL BAR GRAPH, STATEWISE CONFIRMED CASES )
statewise_data = data['statewise']
states = [entry['state'] for entry in statewise_data[1:]]  # Skip the first entry as it is the total for India
confirmed_cases = [int(entry['confirmed']) for entry in statewise_data[1:]]
```

```python
plt.figure(figsize=(12, 6))
plt.barh(states, confirmed_cases)
plt.title('State-wise Confirmed COVID-19 Cases in India')
plt.xlabel('Number of Confirmed Cases')
plt.ylabel('States')
plt.tight_layout()
plt.show()
```
