```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Titanic Dataset.csv")
```

```python
data.head()
```

```python
cleaned_data = data.dropna(subset=['age', 'fare'])
```

```python
print(cleaned_data.head())
```

```python
plt.scatter(cleaned_data['age'], cleaned_data['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
plt.show()
```

```python
X = cleaned_data[['age']]
y = cleaned_data['fare']
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
predictions = model.predict(X)
```

```python
plt.scatter(cleaned_data['age'], cleaned_data['fare'], label='Data')
plt.plot(cleaned_data['age'], predictions, color='red', label='Linear Regression')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Linear Regression of Fare based on Age')
plt.legend()
plt.show()
```
