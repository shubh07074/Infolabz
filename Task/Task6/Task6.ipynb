```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import cv2
import os
```

```python
#1. Print/display First image using tensorflow ( image ) [first image of laptop from training data.]

image_dir = "imagedata/imagedata/training/laptop"

# Get the first image file in the directory
first_image_file = os.listdir(image_dir)[0]
first_image_path = os.path.join(image_dir, first_image_file)

# Load the first image
img = image.load_img(first_image_path)

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

```

```python
#2. Print the size of image
print(f"Size of the image: {img.size}")
```

```python
#3. Print Ndimension
img_cv2 = cv2.imread(first_image_path)
print(f"Number of dimensions: {img_cv2.ndim}")
```

```python
#4. Print grayscale image
# Convert the image to grayscale
gray_image = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2GRAY)

# Check if the grayscale conversion is successful
if gray_image is None:
    print("Failed to convert image to grayscale")
else:
    print("Successfully converted image to grayscale")
    print(f"Grayscale image shape: {gray_image.shape}")

# Display the grayscale image using Matplotlib
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#5. Print all images
image_files = os.listdir(image_dir)
for img_file in image_files:
    img_path = os.path.join(image_dir, img_file)
    imgall = image.load_img(img_path)
    
    plt.imshow(imgall)
    plt.axis('off')
    plt.show()
```
